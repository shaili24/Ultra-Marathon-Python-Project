```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event']== 2020) & (df['Athlete country'] == 'USA')]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
data2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) &
 (df['Year of event']== 2020) &
  (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
data2.head()
```

```python

data2['Event name'].str.split('(').str.get(0)
```

```python
# Remove (USA) from event name
data2['Event name'] =data2['Event name'].str.split('(').str.get(0)
```

```python
data2.head()
```

```python
#clean up age

data2['athlete_age'] = 2020 - data2['Athlete year of birth']
```

```python
data2.head()
```

```python
# Remove h from athlete performance
data2['athlete_performance'] = data2['athlete_performance'].str.split(' ').str.get(0)
```

```python
# deleting the coulmns = Athlete club, Athlete Country, Athlete year of birth, Athlete age category

data2 = data2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
data2.head()
```

```python
#clean null values

data2.isna().sum()
```

```python
data2[data2['athlete_age'].isna() == 1]
```

```python
data2 = data2.dropna()
```

```python
data2.isna().sum()
```

```python
# check for duplicates

data2[data2.duplicated() == True]
```

```python
# reset index

data2.reset_index(drop=True)
```

```python
# changing dtypes

data2.dtypes
```

```python
data2['athlete_age'] = data2['athlete_age'].astype(int)
```

```python
data2['athlete_avg_speed'] = data2['athlete_avg_speed'].astype(float)
```

```python
data2.dtypes
```

```python
# rename names
```

```python
data2 = data2.rename(columns= {
'Year of event' : 'year',
'Event dates' : 'race_day',
'Event name' : 'race_name',
'Event distance/length' : 'race_length',
'Event number of finishers' : 'race_number_of_finishers',
'Athlete performance' : 'athlete_performance',
'Athlete gender' : 'athlete_gender',
'Athlete average speed' : 'athlete_avg_speed',
'Athlete ID' :'athlete_id'
})
```

```python
#Difference in speed for 50km, 50 mi male to female

data2.groupby(['race_length', 'athlete_gender'])['athlete_avg_speed'].mean()
```

```python
# what age groups are the best in 50m Race (20+ race min)

data2.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count> 19').head(15)
```

```python
# what age groups are the worst in 50m Race (10+ race min)

data2.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count> 9').head(15)
```

```python
# Seasons for the data -> Slower in summer than winter?

data2['month'] = data2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
data2['race_season'] = data2['month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x> 5 else 'Spring' if x>2 else 'Winter')
```

```python
data2.head(5)
```

```python
data2.groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# For 50 mile only

data2.query('race_length == "50mi"').groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
